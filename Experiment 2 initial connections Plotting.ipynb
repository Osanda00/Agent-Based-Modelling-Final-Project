```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("experiment_1_1.csv")
```

```python
df
```

```python
df.columns
```

```python
averages = df.groupby(['average connections']).mean()
```

```python
averages
```

```python
averages.to_csv('averaged_results_Experiment 1.csv')
```

```python
df_avg = pd.read_csv('averaged_results_Experiment 1.csv')
```

```python
df_avg
```

```python
x = np.arange(0,30,1)
z1 = df_avg['idea_0'].values
z2 = df_avg['idea_1'].values
z3 = df_avg['idea_2'].values
```

```python
plt.figure(figsize=(10, 6))

# Plot each z value against x
plt.plot(x, z1, label='Neutral')
plt.plot(x, z2, label='Idea 1 holders')
plt.plot(x, z3, label='Idea 2 holders')

# Add labels and title
plt.xlabel('No. of avarage connections')
plt.ylabel('Propotion')
plt.title('Propotion of idea holders vs avarage connections')
plt.legend()

# Show the plot
plt.grid(True)
plt.savefig('my_plot.png') 
plt.show()
```
